```python
df = pd.read_csv('grand_slams_15.csv', index_col = 0)
```

```python
df.columns
```

```python
winner_df = df[['winner_seed', 'winner_entry',
       'winner_name', 'winner_hand', 'winner_ht', 'winner_ioc', 'winner_age',
       'winner_rank', 'winner_rank_points', 'w_ace', 'w_df', 'w_svpt', 'w_1stIn', 'w_1stWon',
       'w_2ndWon', 'w_SvGms', 'w_bpSaved', 'w_bpFaced']]
winner_df['wins'] = 1
winner_df['seeded'] = winner_df['winner_seed'].notnull()
```

```python
player_df = winner_df.groupby('winner_name').aggregate({'wins': 'sum', 'seeded' : 'mean', 'w_ace' :'mean', 'w_df' :'mean', 'w_svpt' :'mean', 'w_1stIn' :'mean',
                                            'w_1stWon' :'mean', 'w_2ndWon' :'mean', 
                                            'w_SvGms' :'mean'})
```

```python
player_df
```

```python
loser_df = df[['loser_seed',
       'loser_entry', 'loser_name', 'loser_hand', 'loser_ht', 'loser_ioc',
       'loser_age', 'loser_rank', 'loser_rank_points','l_ace', 'l_df',
       'l_svpt', 'l_1stIn', 'l_1stWon', 'l_2ndWon', 'l_SvGms'
              ]]
loser_df['losses'] = 1
loser_df['seeded'] = loser_df['loser_seed'].notnull()
```

```python
player_losses = loser_df.groupby('loser_name').aggregate({'losses': 'sum', 'seeded' : 'mean', 'l_ace' :'mean', 'l_df' :'mean', 'l_svpt' :'mean', 'l_1stIn' :'mean',
                                            'l_1stWon' :'mean', 'l_2ndWon' :'mean'})
```

```python
player_losses.reset
```

```python
player_df = pd.merge(player_df, player_losses, 'inner', left_index = True, right_index=True)
```

```python
player_df['total_games'] = player_df['wins'] + player_df['losses']
```

```python
player_df = player_df[player_df['total_games'] > 10]
```

```python
player_df['per_seeded'] = ((player_df['wins'] * player_df['seeded_x']) + (player_df['losses'] * player_df['seeded_y'])) / player_df['total_games']
```

```python
player_df['win_percentage'] = player_df['wins'] / player_df['total_games']
```

```python
diff_col_names = []
for i in ['w_ace', 'w_df', 'w_1stIn', 'w_1stWon',
       'w_2ndWon']:
    diff_col_names.append(i.split('_')[1] + '_' + 'diff')
```

```python
for i in diff_col_names:
    attr = i.split('_')[0]
    w_attr = 'w_' + attr
    l_attr = 'l_' + attr
    player_df[i] = player_df[w_attr] - player_df[l_attr]
```

```python
player_df['df_diff'] = player_df['df_diff'] * -1
```

```python
player_df
```

```python
tot_cols_name = []
for i in ['w_ace', 'w_df', 'w_1stIn', 'w_1stWon',
       'w_2ndWon']:
    tot_cols_name.append(i.split('_')[1] + '_' + 'tot')
for i in tot_cols_name:
    attr = i.split('_')[0]
    w_attr = 'w_' + attr
    l_attr = 'l_' + attr
    player_df[i] = ((player_df['wins'] * player_df[w_attr]) + (player_df['losses'] * player_df[l_attr])) / (player_df['total_games'])
```

```python
clus_feat = ['ace_diff', 'df_diff', '1stIn_diff', '1stWon_diff',
       '2ndWon_diff', 'ace_tot', 'df_tot', '1stIn_tot', '1stWon_tot',
       '2ndWon_tot']
clus_df = player_df[clus_feat]
```

```python
clus_df['df_tot'] = -1 * clus_df['df_tot'] # we want the more positive number to signify better performance
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
clus_df = pd.DataFrame(mms.fit_transform(clus_df), columns = clus_df.columns, index = clus_df.index )
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 3, random_state = 1)
model.fit(clus_df)
player_df['cluster'] = model.labels_
clus_df['cluster'] = model.labels_
```

```python
summary_df = clus_df.groupby('cluster').mean()
summary_df = summary_df.rename(columns = {'ace_diff':'Avg Ace Differential', 'df_diff':'Avg Double Fault Differential',
                            '1stIn_diff':'Avg 1st Service In Differential', 
                             '2ndWon_diff':'Avg 2nd Service won Differential', 
                             '1stWon_diff':'Avg 1st Service won Differential', 
                            'ace_tot': 'Avg Total Aces in a Match ', 
                            'df_tot': 'Avg Total Double Faults in a Match', 
                            '1stIn_tot': 'Avg Total 1st Serves In', 
                            '1stWon_tot': 'Avg Total 1st Serves Won', 
                            '2ndWon_tot': 'Avg Total 2nd Serves Won'}, errors ='raise').transpose()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (16,8))
sns.heatmap(summary_df, annot = True, vmin =0 , vmax = 1, cmap = 'Greens')
plt.title('Each Cluster Broken Down By Scaled Score (0 - 1) Of Features', fontdict = {'fontsize' : 16, 'fontweight': 10})
plt.ylabel('Feature', fontdict = {'fontsize' : 15})
plt.xlabel('Cluster', fontdict = {'fontsize' : 15})
```

## Summary of each cluster

Group 0: The average server, that tend to be more aggresive with their serves (highest double faults and least 1st serves that go in) - nothing special but not as bad in terms of points returned as the safe servers

Group 1: The best, most successful servers - striking a balance between aggressive and defensive serves

Group 2: The safe servers (Low double faults, aces, and high percentage of first serves in), but they win the least points out of everyone
